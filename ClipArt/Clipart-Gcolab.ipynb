*   this script opens cartoonized_image folder and converts images to PNG
*   Only accepts JPG JPEG and PNG, this will IGNORE ANY OTHER IMAGE FORMAT
*   after convertion THIS WILL DELETE THE ORIGINAL IMAGE only png image format will be in the folder.

```python
import os
from rembg import remove
from PIL import Image
from tqdm import tqdm
import time

#automatically finds the cartoonized_images folder saves the file path to input_dir
CARTOONIZED_PATH = !find -name "cartoonized_images"
input_dir = CARTOONIZED_PATH[0]
output_dir = CARTOONIZED_PATH[0] #comment this if you want to use a different output folder

# uncomment this and Define the desired output directory
#output_dir = '/content/drive/MyDrive/FIBEI GALLERY MAIN TRACKER THINGS CONVERTED UPLOAD/JASPER'

files = os.listdir(input_dir)

#covnert jpg to png first this is because when u use remove(filename.jpg) the background is just black
for filename in tqdm(files, desc="Converting images to png", unit='images'):
    if filename.endswith('.jpg') or filename.endswith('.JPG'):
      jpg_filepath = os.path.join(input_dir, filename)
      #input_image = Image.open(input_path)
      #rgb_im = input_image.convert('RGB')
      #rgb_im.save(input_path)
      img = Image.open(jpg_filepath)

      # Convert the image to PNG format
      png_filename = filename[:-4] + ".png"
      #png_filepath = os.path.join(input_dir, png_filename) #store in the same folder
      png_filepath = os.path.join(output_dir, png_filename) #store in another folder 
      img.save(png_filepath, "PNG")
      os.remove(jpg_filepath) #delete the .jpg file

    elif filename.endswith('.jpeg') or filename.endswith('.JPEG'):
      jpg_filepath = os.path.join(input_dir, filename)
      #input_image = Image.open(input_path)
      #rgb_im = input_image.convert('RGB')
      #rgb_im.save(input_path)
      img = Image.open(jpg_filepath)

      # Convert the image to PNG format
      png_filename = filename[:-5] + ".png"
      #png_filepath = os.path.join(input_dir, png_filename)#store in the same folder
      png_filepath = os.path.join(output_dir, png_filename) #store in another folder 
      img.save(png_filepath, "PNG")
      os.remove(jpg_filepath) #delete the .jpg file
      #print(png_filepath)
```

# STEP 9 BY BATCH IMAGE BACKGROUND REMOVAL (TRANSPARENT)




*   This script will remove the background of the images from cartoonzied_images folder
*   You can automatically save it to a designated folder follow this step
  1.   open the file explorer, (left side)
  2.   go to the folder you want to save the images to
  3.   right click on the folder and select copy path
  4.   go to variable output_dir and replace the value

```python
import os
from rembg import remove
from PIL import Image
from tqdm import tqdm
import time

#automatically finds the cartoonized_images folder saves the file path to input_dir
CARTOONIZED_PATH = !find -name "cartoonized_images"
input_dir = CARTOONIZED_PATH[0]

# Define the input desired output directory
output_dir = '/content/drive/MyDrive/Colab Notebooks/ClipArt/test_code/OUTPUT' #copy path to ouput folder

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

files = os.listdir(input_dir)

# Loop through all the image files in the input directory
for filename in tqdm(files, desc="Removing background", unit='images'):
    if filename.endswith('.png'):
      
        # Load the image from the input directory
        input_path = os.path.join(input_dir, filename)
        input_image = Image.open(input_path)

        # Remove the background using the rembg library
        output_image = remove(input_image)

        # Save the transparent image to the output directory
        output_path = os.path.join(output_dir, filename)
        #output_image.save(output_path)
        #rgb_im = output_image.convert('RGB') #converting from .jpg (RGBA) to .png (RGB) does not work
        output_image.save(output_path)

print('All images processed successfully.')

```
