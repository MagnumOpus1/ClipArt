```python
import csv
import os
import gdown


# Variables used in the program
filename = 'link.csv'

def downloader(link):
    """
        Downloader function
    """
    # Get the file id from the sharable link
    # It is probably second from the last
    file_id =  link.split('/')[-2]
    prefix = 'https://drive.google.com/uc?/export=download&id='
    # Download the specific image
    #gdown.download(prefix+file_id)
    gdown.download(prefix+file_id, output = '/content/drive/MyDrive/Colab Notebooks/ClipArt/test_code/test_images')
def main():
    """
        Main function
    """
    csv_file = os.path.join(os.getcwd(), filename)
    #counter = 1
    # Download the images from the links in the csv file
    with open(csv_file, 'r') as file:
        links = csv.reader(file)
        for link in links:
            # Note, get the 0th element since link contains a list with the link as a single element
            image_link = link[0]
            downloader(image_link)
            
            #counter += 1
    
if __name__ == "__main__":
    main()
```
